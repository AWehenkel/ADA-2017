```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import matplotlib.dates as pltdates
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("final_merge.csv")
```

```python
df_c = df.copy()
df_c = df_c.loc[df_c['year']!=0]

year_plot = df_c.groupby('year').size().plot(kind='bar', figsize=(16,12))
ax = plt.gca()
for label in ax.get_xticklabels():
    label.set_visible(False)
for label in ax.get_xticklabels()[2::5]:
    label.set_visible(True)
```

We can see, as expected, that the dataset doesn't have many songs before 1990. It also stops after 2010 (when the dataset got created).

```python
def plot_by(df, idx, column, ignore_idx=None):
    df_c = df.copy()
    df_c = df_c[[idx,column]]
    if ignore_idx!=None:
        df_c = df_c.loc[df_c[idx]!=ignore_idx]
    df_c.groupby([idx]).mean().plot()
    
def plot_by_year(df, column):
    plot_by(df,'year',column,0)
    
def plot_heatmap_by(df, idx, column, ignore_idx=None):
    df_c = df.copy()
    df_c = df_c[[idx,column]]
    if ignore_idx!=None:
        df_c = df_c.loc[df_c[idx] != 0]
    if df[column].dtype == np.float or df[column].dtype == np.float64:
        bins = np.linspace(df[column].min(),df[column].max(),20)
        df_c[column] = pd.cut(df_c[column],bins)
    df_c = df_c.dropna()
    df_count = pd.DataFrame(df_c.groupby([idx, column]).size().rename('count'))
    df_c = df_c.join(df_count, on=[idx,column])
    df_c = df_c.reset_index().pivot_table(index=idx, columns=column, values='count', aggfunc='mean')
    plt.clf()
    fig = plt.gcf()
    fig.set_size_inches(16,12)
    sns.heatmap(df_c, figure=fig)
    plt.show()

def plot_heatmap_by_year(df, column):
    plot_heatmap_by(df, 'year', column, 0)
    
def plot_for_year(df, column, year):
    df_c = df.copy()
    df_c = df_c[df_c['year']==year]
    df_c = df_c[[column]]
    plt.clf()
    fig = plt.gcf()
    axes = plt.gca()
    fig.suptitle(column+" distribution for year "+str(year))
    fig.set_size_inches(16,12)
    axes.set_xlabel(column)
    axes.set_ylabel(year)
    sns.stripplot(df_c, jitter=0.4)
```
